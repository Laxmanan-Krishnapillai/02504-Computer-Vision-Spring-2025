## Exercise 1.1
Convert 2D points from homogeneous to inhomogeneous coordinates.

```python

def from_hom(p):
    return [p[0]/p[2], p[1]/p[2]]
points=[[1,2,1],[4,2,2],[6,4,-1],[5,3,0.5]]
for i,p in enumerate(points,1):
    q=from_hom(p)
    print(f"q{i} = {q}")
```

## Exercise 1.2
Convert 3D points from homogeneous to inhomogeneous coordinates.

```python

def from_hom3(P):
    return [P[0]/P[3], P[1]/P[3], P[2]/P[3]]
points=[[1,10,-3,1],[2,-4,1.1,2],[0,0,-1,10],[-15,3,6,3]]
for i,P in enumerate(points,1):
    Q=from_hom3(P)
    print(f"Q{i} = {Q}")
```

## Exercise 1.5
Find the intersection of two lines.

```python

def cross(u,v):
    return [u[1]*v[2]-u[2]*v[1], u[2]*v[0]-u[0]*v[2], u[0]*v[1]-u[1]*v[0]]
l0=[1,1,-1]
l1=[-1,1,-3]
q=cross(l0,l1)
qi=[q[0]/q[2], q[1]/q[2]]
print('Intersection q_I =', qi)
```

## Exercise 1.7
Compute distances from line $l=[1/\sqrt{2},1/\sqrt{2},-1]^T$ to given points.

```python

l=[1/math.sqrt(2),1/math.sqrt(2),-1]
pts=[[0,0,1],[math.sqrt(2),math.sqrt(2),1],[math.sqrt(2),math.sqrt(2),4]]
for i,p in enumerate(pts,1):
    num=abs(l[0]*p[0]+l[1]*p[1]+l[2]*p[2])
    den=math.sqrt(l[0]**2+l[1]**2)
    print(f"distance p{i} = {num/den}")
```

## Exercise 1.11
Define a function `box3d` that generates points on a 3D box.

```python

import itertools

def box3d(n=16):
    pts=[]
    lin=[-0.5 + i*(1.0/(n-1)) for i in range(n)]
    edges=[(-0.5,-0.5),(-0.5,0.5),(0.5,0.5),(0.5,-0.5),(0,0)]
    for i,j in edges:
        for x in lin:
            pts.append([i,x,j])
            pts.append([x,i,j])
            pts.append([i,j,x])
    return pts
pts=box3d(4)
print('Generated', len(pts), 'points')
```

## Exercise 1.12
Implement $\Pi$ and $\Pi^{-1}$.

```python

def Pi(ph):
    return [c/ph[-1] for c in ph[:-1]]

def PiInv(p):
    return p + [1]
print(PiInv(Pi([2,4,2])))
```

## Exercise 1.13
Project 3D points using a pinhole camera model.

```python

def matmul(A,B):
    m=len(A); n=len(B[0]); k=len(B)
    return [[sum(A[i][t]*B[t][j] for t in range(k)) for j in range(n)] for i in range(m)]

def projectpoints(K,R,t,Q):
    Qh=[q+[1] for q in Q]
    Rt=[r+[t[i]] for i,r in enumerate(R)]
    M=matmul(K,Rt)
    qs=[matmul(M,[q])[0] for q in Qh]
    return [Pi(q) for q in qs]
K=[[1,0,0],[0,1,0],[0,0,1]]
R=[[1,0,0],[0,1,0],[0,0,1]]
t=[0,0,4]
Q=[[0,0,0]]
print(projectpoints(K,R,t,Q))
```
